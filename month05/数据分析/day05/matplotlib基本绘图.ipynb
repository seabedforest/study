```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#绘制折线图
x=np.array([1,2,3,4,5,6])
y=np.array([12,33,74,51,8,80])
plt.plot(x,y)
plt.show()
```

```python
#绘制正弦，余弦函数
#从【-π，π】拆出200个点
x = np.linspace(-np.pi,np.pi,200)
sinx=np.sin(x)  #矢量化方法  可以针对一组数据操作
plt.plot(x,sinx)
```

```python
#绘制【-π，π】区间的: y=cos(x)/2
cosx=np.cos(x)/2
plt.plot(x,cosx)
```

```python
plt.plot(x,sinx,linewidth=2,linestyle='--',color='dodgerblue',alpha=0.9)
plt.plot(x,cosx,linewidth=2,linestyle='--',color='orangered',alpha=0.9)
```

```python
plt.plot(x,sinx,linewidth=2,linestyle='--',color='dodgerblue',alpha=0.9)
plt.plot(x,cosx,linewidth=2,linestyle='--',color='orangered',alpha=0.9)
#设置坐标轴的范围
plt.xlim(0,np.pi)
plt.ylim(-1,1)
```

```python
plt.plot(x,sinx,linewidth=2,linestyle='--',color='dodgerblue',alpha=0.9)
plt.plot(x,cosx,linewidth=2,linestyle='-',color='orangered',alpha=0.9)
#设置x轴刻度文本
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\frac{\pi}{2}$','0','π/2','π'])
plt.show()
```

```python
plt.plot(x,sinx,linewidth=2,linestyle='--',color='dodgerblue',alpha=0.9)
plt.plot(x,cosx,linewidth=2,linestyle='-',color='orangered',alpha=0.9)
#设置坐标轴 修改成平面直角坐标系
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
plt.yticks([-1,-0.5,0.5,1])
#绘制特许点 [π/2,1][π/2,0]
plt.scatter([np.pi/2,np.pi/2],[0,1],marker='p',s=80,edgecolor='green',facecolor='red',zorder=3)
#备注
plt.annotate(
    r'$[\frac{\pi}{2},1]$',
    xycoords='data',
    xy=(np.pi/2,1),
    textcoords='offset points',
    xytext=(30,20),
    fontsize=14,
    arrowprops=dict(
        arrowstyle='->',
        connectionstyle='angle3')
)
plt.show()
```

```python
#自定义窗口内容
plt.figure('FA',facecolor='gray')
plt.title('FA Title')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.grid(linestyle='solid')
plt.figure('FB',facecolor='lightgray')
plt.title('FB Title')
plt.show()
```

```python
#子图
#绘制3行3列的子图
plt.subplot(3,3,1)
plt.text(0.5,0.5,'1',size=36,va='center',ha='center')
plt.xticks([])
plt.yticks([])
#plt.tight_layout()
plt.show()
```

```python
plt.figure('九宫格',facecolor='blue')
for i in range(1,10):
    plt.subplot(3,3,i)
    plt.text(0.5,0.5,i,size=36,va='center',ha='center')
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
#网格式布局
import matplotlib.gridspec as mg
gs = mg.GridSpec(3,3)
#合并第一行的前两列
plt.subplot(gs[0:2])
plt.text(0.5,0.5,1,size=30,va='center',ha='center')
plt.xticks([])
plt.yticks([])

plt.subplot(gs[:2,-1])
plt.text(0.5,0.5,2,size=30,va='center',ha='center')
plt.xticks([])
plt.yticks([])

plt.subplot(gs[1,1])
plt.text(0.5,0.5,3,size=30,va='center',ha='center')
plt.xticks([])
plt.yticks([])

plt.subplot(gs[1:,-3])
plt.text(0.5,0.5,4,size=30,va='center',ha='center')
plt.xticks([])
plt.yticks([])

plt.subplot(gs[-1,1:])
plt.text(0.5,0.5,5,size=30,va='center',ha='center')
plt.xticks([])
plt.yticks([])

plt.show()
```

```python
#散点图
n=200
x=np.random.normal(175,5,n)
y=np.random.normal(65,10,n)
#plt.scatter(x,y,s=60,color='dodgerblue',edgecolor='orangered',label='Samples')
plt.scatter(x,y,s=60,c=x,cmap='brg',label='Samples')
plt.legend()
```

```python
#绘制12个月苹果的销量
x=np.arange(12)
apples=np.random.normal(80,15,12)
plt.bar(x-0.2,apples,0.4,color='orangered',label='apples',alpha=0.8)
oranges=np.random.normal(80,15,12)
plt.bar(x+0.2,oranges,0.4,color='dodgerblue',label='apples',alpha=0.8)
plt.legend()
plt.xticks(x,['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.show()
```

```python
#分析性别对学习成绩的影响
import pandas as pd
data=pd.read_csv('../day04/学生考试表现数据/StudentsPerformance.csv')
r=data.pivot_table(index='gender')
r
```

```python
x=np.arange(3)
female = r.loc['female']
male=r.loc['male']
plt.bar(x-0.2,female,0.3,color='orangered',label='Female')
plt.bar(x+0.2,male,0.3,color='dodgerblue',label='Male')
plt.xticks(x,r.columns)
plt.legend()
plt.show()
```
