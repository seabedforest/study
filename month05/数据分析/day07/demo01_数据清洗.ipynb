```python
data = pd.DataFrame({'id':[1,1,2,2],'name':['zs','ls','zs','ls']})
data
```

```python
#检测与处理重复值
data.drop_duplicates(subset='id')
```

```python
ratings=pd.read_json('ratings.json')
ratings.isnull().sum().sum() #检测缺失值
```
