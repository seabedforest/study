```python
ratings=pd.read_json('../day03/ratings.json')
```

```python
def func(row):
    #寻找一行中的NAN,赋值为均值
    row[pd.isna(row)]=row.mean()
    return row
ratings.apply(func,axis=1)
```

```python
data=pd.read_csv('../day03/二手房数据/house.csv')
data
#求所有房源数据平米数的均值
def func(item):
    if '平米' in item:
        return float(item[:-2])
    else:
        return None
data['square'].apply(func).median()
```

```python
data['square'].apply(func).mean()
```
