```python
#二项分布
#某人投篮命中率为0.3，投10次，进5个球的概率
r=np.random.binomial(10,0.3,100)
(r==5).sum() / 100
for i in range(11):
    print(i,":",(r==i).sum() / 100)
```

```python
a=np.random.binomial(3,0.6,100)
(a==0).sum()/100
```

```python
#超几何分布
#从6个好苹果、4个坏苹果中抽取3个苹果，返回好苹果的数量（执行10次)
r= np.random.hypergeometric(6,4,3,10)
for i in range(4):
    print(i,":",(r==i).sum()/10)
```

```python
import matplotlib.pyplot as plt
values = [15, 13.3, 8.5, 7.3, 4.62, 51.28]
spaces = [0.05, 0.01, 0.01, 0.01, 0.01, 0.01]
labels = ['Java', 'C', 'Python', 'C++', 'VB', 'Other']
plt.axis('equal')
colors = ['dodgerblue', 'orangered', 'limegreen', 'violet', 'gold','blue']
plt.pie(values,spaces,labels,colors,'%.1f%%',shadow=True,startangle=0)
plt.legend()
plt.show()
```

```python
import pandas as pd
data=pd.read_csv('../day04/学生考试表现数据/StudentsPerformance.csv')
mdata=data.sort_values(by='math score',ascending=False).head(100)
#绘制饼状图
values = mdata['gender'].value_counts()
spaces=[0.05,0.01]
labels=mdata['gender'].value_counts().index
colors=['dodgerblue','orangered']
plt.axis('equal')
plt.pie(values,spaces,labels,colors,'%.1f%%',shadow=True,startangle=90)
plt.legend()
#plt.show()
```
