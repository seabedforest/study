```python
import numpy as np
import sklearn.preprocessing as sp
```

```python
#归一化（每一行绝对值之和为1）
data=np.array([[20,10,5],[4,2,1],[15,11,13]])
r=sp.normalize(data,norm='l1')
print(r)
print(r.sum(axis=1))
```

```python
r2=sp.normalize(data,norm='l2')
print(r2)
print((r2**2).sum(axis=1))
```

```python
#二值化
bin=sp.Binarizer(threshold=10)
r=bin.transform(data)
r
```

```python
#独热编码
data= np.array([[1,3,2],[7,5,4],[1,8,6],[7,3,9]])
ohe=sp.OneHotEncoder()
r=ohe.fit_transform(data)
print(r)
```

```python
#标签编码
xueli = np.array(['本科','专科','硕士','本科','本科','本科','本科','专科','博士'])
lbe = sp.LabelEncoder()
r=lbe.fit_transform(xueli)
r
```

```python
#需要反查词典，得到编码原文本
lbe.inverse_transform(r)
```
