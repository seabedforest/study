```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#模拟一组房价数据，实现房价预测
#[56,350] [62,420] [82,530] [108,680] [75,480] [92,600]
#预测[120:?]
x=np.array([56,62,82,108,75,92])
y=np.array([350,420,530,680,480,600])


plt.grid(linestyle=':')
plt.scatter(x,y,marker='s',s=60,color='dodgerblue')
```

```python
w0,w1=1,1 #模型参数
lrate=0.00001 #学习率
times=1000  #迭代次数
#执行梯度下降流程，不断更新w0与w1，找到最优线性模型
for i in range(times):
    #通过偏导公式，计算w0与w1方向上的偏导数
    d0=(w0+w1*x-y).sum()
    d1=(x*(w0+w1*x-y)).sum()
    #更新w0与w1
    w0=w0-lrate*d0
    w1=w1-lrate*d1
    
plt.grid(linestyle=':')
plt.scatter(x,y,marker='s',s=60,color='dodgerblue')
pred_y = w0+w1*x
plt.plot(x,pred_y,color='orangered',label='Regression Line')
plt.legend()
plt.show()
```

```python
import pandas as pd
```

```python
data=pd.read_csv('Salary_Data.csv')
data.plot.scatter(x='YearsExperience',y='Salary')
```

```python
data.plot(x='YearsExperience',y='Salary')
```

```python
#使用sklearn内置的线性回归API,训练模型
import sklearn.linear_model as lm
model = lm.LinearRegression()
x,y=data.loc[:,:'YearsExperience'],data['Salary']
model.fit(x,y)
pred_y=model.predict(x)
data['pred_y']=pred_y
ax=data.plot(x='YearsExperience',y='pred_y')
data.plot.scatter(x='YearsExperience',y='pred_y',ax=ax)
```
