```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([0.5,0.6,0.8,1.1,1.4])
y=np.array([5.0,5.5,6.0,6.8,7.0])


plt.grid(linestyle=':')
plt.scatter(x,y,marker='s',s=60,color='dodgerblue')
```

```python
w0,w1=1,1 #
lrate=0.01 #
times=1000  #
w0s,w1s,losses,epoches=[],[],[],[]
#w0w1
for i in range(times):
    #w0w1loss
    loss = ((w0+w1*x-y)**2).sum()/2
    print('{:4}:w0={:.8f},w1={:.8f},loss={:.8f}'.format(i+1,w0,w1,loss))
    #
    w0s.append(w0)
    w1s.append(w1)
    losses.append(loss)
    epoches.append(i+1)
    #w0w1
    d0=(w0+w1*x-y).sum()
    d1=(x*(w0+w1*x-y)).sum()
    #w0w1
    w0=w0-lrate*d0
    w1=w1-lrate*d1
    
```

```python
#
plt.title('Traning Progress')
plt.subplot(3,1,1)
plt.grid(linestyle=':')
plt.plot(epoches,w0s,color='dodgerblue',label='w0')
plt.legend()

plt.subplot(3,1,2)
plt.grid(linestyle=':')
plt.plot(epoches,w1s,color='dodgerblue',label='w1')
plt.legend()

plt.subplot(3,1,3)
plt.grid(linestyle=':')
plt.plot(epoches,losses,color='orange',label='loss')
plt.legend()
plt.tight_layout()   #
```

```python
plt.grid(linestyle=':')
plt.scatter(x,y,marker='s',s=60,color='dodgerblue')
pred_y = w0+w1*x
plt.plot(x,pred_y,color='orangered',label='Regression Line')
plt.legend()
plt.show()
```
