```python
import numpy as np
import sklearn.datasets as sd
```

```python
#
data = sd.load_files('20news/',encoding='latin1')
data
```

```python
print(data['target_names']) #
print(data['data'][0])   #
print(data['target'][0])  #
```

```python
print(data['target'][1])
```

```python
#tfidf  
import sklearn.feature_extraction.text as ft
import sklearn.model_selection as ms
import sklearn.linear_model as lm
import sklearn.metrics as sm
```

```python
cv=ft.CountVectorizer()
bow=cv.fit_transform(data['data'])
tt = ft.TfidfTransformer()
tfidf=tt.fit_transform(bow)
tfidf.shape
```

```python
train_x,test_x,train_y,test_y=ms.train_test_split(tfidf,data['target'],test_size=0.1,random_state=7)

#
model = lm.LogisticRegression()
# 
import sklearn.naive_bayes as nb
model = nb.GaussianNB()
model = nb.MultinomialNB()
model.fit(train_x,train_y)
pred_test_y=model.predict(test_x)
print(sm.classification_report(test_y,pred_test_y))
```

```python
# 
test_data = [
    'The curveballs of right handed pitchers tend to curve to the left',
    'Caesar cipher is an ancient form of encryption',
    'This two-wheeler is really good on slippery roads']
#
bow = cv.transform(test_data)
tfidf=tt.transform(bow)
pred_test_y = model.predict(tfidf)
print(pred_test_y)
print(data['target_names'])
```
